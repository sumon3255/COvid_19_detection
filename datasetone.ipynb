```python
import pandas as pd
import os
import shutil
```

```python
File_Path = "chestXray/metadata.csv"
Img_Path = "chestXray/images"
```

```python
df = pd.read_csv(File_Path)
```

```python
df.shape
```

```python
df.tail()
```

```python
Targer_dir = "my_dataset/Covid"
if not os.path.exists(Targer_dir):
    os.mkdir(Targer_dir)
    print("Folder Created")
```

```python
cnt = 0
for (i,row) in df.iterrows():
    if row["finding"] == "COVID-19" and row["view"] == "PA":
        filename = row["filename"]
        image_path = os.path.join(Img_Path , filename )
        img_copy_path = os.path.join(Targer_dir, filename )
        
        shutil.copy2(image_path, img_copy_path)
        cnt = cnt+1
print(cnt)
```
